f(x) = sin(x)^2 - cos(x)^2
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0


```python
import sympy as sp 
import numpy as np
```

```python
x = sp.Symbol('x')
f = sp.sin(x)**2 - sp.cos(x)**2
fSol = sp.solve(f, x)
print (f'Корни данного уравнения: {fSol}')
```

```python
fDif = sp.diff(f)
fDifSolInc = sp.solve(fDif > 0)
print (f'Функция возрастает на интервале: {fDifSolInc}')
```

```python
fDifSolDec = sp.solve(fDif < 0)
print (f'Функция убывает на интервале: {fDifSolDec}')
```

```python

p1 = sp.plotting.plot(f, title = 'График функции', legend=True, show=False, xlim=(-5, 5))
p1.show()
```

```python
peaks = sp.solve(fDif, x)
print (f'Вершины графика: {peaks}')
```

```python
fPos = sp.solve(f > 0)
print (f'Функция > 0 на промежутках: {fPos}')
```

```python
fNeg = sp.solve(f < 0)
print (f'Функция < 0 на промежутках: {fNeg}')
```
